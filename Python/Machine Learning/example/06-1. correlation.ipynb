```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
iris = load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
col_names = [
             'sepal_length',  # 꽃받침의 길이
             'sepal_width' ,  # 꽃받침의 폭
             'petal_length',  # 꽃잎의 길이
             'petal_width' ,  # 꽃잎의 폭
             'target'
            ]
```

```python
iris_df = pd.DataFrame(data= np.c_[iris.data, iris.target], columns= col_names)
iris_df.target = iris_df.target.map({0: "setosa", 1: "versicolor", 2: "virginica"})
```

```python
iris_df
```

```python
plt.figure(figsize=(8,8))
for i in range(0,4):    
    plt.subplot(2,2,i+1)
    sns.distplot(iris_df.iloc[:,i])
plt.tight_layout()
plt.show()
```

```python
iris_corr = iris_df.corr()
iris_corr
```

```python
sns.pairplot(iris_df)
plt.show()
```

```python
sns.pairplot(iris_df, kind='reg')
plt.show()
```
