```python
import matplotlib.pyplot as plt
```

# 기본 그래프 그리기
* plt.plot(data)
* plot에 데이터를 전달하여 그린다.


## y축 데이터로 그리기
* 1차원 리스트, 튜플, 시리즈 데이터를 전달하여 그린다.
* x축은 데이터의 인덱스로 자동 지정된다.

### 리스트 데이터로 그리기

```python
# y축 데이터 [1,5,3,9,7]
data = [1,5,3,9,7]
```

```python
# 그래프 그리기
plt.plot(data)
```

### 시리즈 데이터로 그리기

```python
# pandas import
import pandas as pd
```

```python
# 시리즈 데이터 만들기 ([1,5,3,9,7])
s = pd.Series([1,5,3,9,7])
s
```

```python
# 그래프 그리기 (plot)
plt.plot(s)
```

```python
# 시리즈 데이터의 인덱스 변경하기 : index = [10,20,30,40,50]
ss = pd.Series([1,5,3,9,7], index=[10,20,30,40,50])
ss
```

```python
# 그래프 그리기
plt.plot(ss)
```

```python
# 시리즈 데이터의 인덱스 변경하기 : index = ['a','b','c','d','e']
sss = pd.Series([1,5,3,9,7], index=['a','b','c','d','e'])
sss
```

```python
# 그래프 그리기
plt.plot(sss)
```

## x축, y축 데이터로 그리기
* x축과 y축 데이터의 길이가 같아야 한다.

### 리스트데이터로 그리기

```python
x = [1,5,6,9,10]
y = [1,5,3,9,7]

# 그래프 그리기
plt.plot(x,y)
```

### 데이터프레임 데이터로 그리기

```python
sss = pd.DataFrame({'x':[1,5,6,9,10], 'y':[1,5,3,9,7]})
sss
```

```python
# 그래프 그리기
x = sss['x']
y = sss['y']
plt.plot(x,y)
```

### 시리즈 데이터로 그리기

```python
s = pd.Series([1,5,3,9,7])
s
```

```python
# 시리즈만 전달하여 그리기
plt.plot(s)
plt.show()
```

```python
# x, y 각각 전달하여 그리기
plt.plot(s.index, s)
plt.show()
```
