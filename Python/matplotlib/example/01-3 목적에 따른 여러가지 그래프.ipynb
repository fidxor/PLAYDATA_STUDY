# 라이브러리 임포트
* matplotlib의 pyplot 모듈 사용
* 관용적으로 plt라는 별칭 사용

```python
import matplotlib.pyplot as plt
```

# 시간의 시각화

```python
# 샘플데이터 : 어떤 아이의 나이별 키의 변화
age = [7,8,9,10,11,12,13,14,15]
height = [122.1, 127.9, 133.4, 138.8, 142.7, 151.4, 155.6, 165.0, 169.2]
```

## 선그래프

```python
plt.plot(age, height)
```

# 수량비교 시각화

```python
# 샘플데이터 : 100명 중 혈액형 인원 비교
blood_type=['A','B','O','AB']
count=[45,35,15,5]
```

## 막대 그래프

```python
plt.bar(blood_type,count)
```

## 가로 막대 그래프

```python
plt.barh(blood_type,count)
```

# 비율의 시각화

```python
# 샘플데이터 : 100명 중 혈액형 인원의 비율
blood_type=['A','B','O','AB']
count=[45,35,15,5]
```

## 파이차트
* pie

```python
plt.pie(count, labels=blood_type)
plt.show()
```

# 분포의 시각화

```python
# 샘플데이터 (1~100사이의 랜덤 정수 1000개)
import numpy as np
data = np.random.randint(1,101,1000)
data
```

## 히스토그램

```python
plt.hist(data, bins=100)
plt.show()
```

## 상자수염그래프
* boxplot

```python
plt.boxplot(data)
plt.show()
```

## 바이올린그래프
* violinplot

```python
plt.violinplot(data)
plt.show()
```

# 관계의 시각화

```python
## 샘플데이터 (지불금액에 따른 팁)
import seaborn as sns
tips = sns.load_dataset('tips')[['total_bill','tip']]
tips
```
