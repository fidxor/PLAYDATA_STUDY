```python
import urllib.request # 원격 서버 파일 요청
from bs4 import BeautifulSoup # html 파싱

# 요청할 url
url = 'http://www.naver.com/index.html'
```

```python
# 1. 원격 서버 파일 요청
res = urllib.request.urlopen(url) # web 문서 get 
data = res.read() # text 형태로 읽음
print(data)
```

```python
# 2. source 디코딩
src = data.decode("utf-8") # 한글로 디코딩
print('source')
print(src)
```

```python
# 3. html 파싱
# html 태그를 인식할 수 있도록 의미있는 단위로 분해하고
# 계층적인 트리 구조를 만드는 과정

html = BeautifulSoup(src, 'html.parser') 
print('html 파싱')
print(html)
```

```python
# 4. 태그 내용
a = html.find('a')
print('a tag : ', a)
print('a tag 내용 : ', a.string)
```

```python
#()함수로 추출 
import urllib.request # 원격 서버 파일 요청
from bs4 import BeautifulSoup # html 파싱

# 요청할 url
url = 'http://www.naver.com/index.html'

# 1. 원격 서버 파일 요청
res = urllib.request.urlopen(url) # web 문서 get 
data = res.read() # text 형태로 읽음

# 2. source 디코딩
src = data.decode("utf-8") # 디코딩
print('source')
print(src)


```

```python
# 3. html 파싱
html = BeautifulSoup(src, 'html.parser') # html source 파싱
print('html 파싱')
print(html)

```

```python
# 4. 태그 내용
a = html.find('a')
print('a tag : ', a)
print('a tag 내용 : ', a.string)
```
